Comparing Strings:-

```python
str1 = "apple"
str2 = "banana"

print(str1 == str2) ## Equal to
print(str1 != str2) ## Not equal to
print(str1 > str2)  ## Less than
print(str1 < str2)  ## Greater than
print(str1 >= str2) ## Greater than or equal to
print(str1 <= str2) ## Lesser than or equal to
```

**find() function**--> returns -1 if substring is noty present in the main string

```python
string = "Hello, World!"
index = string.find("World")
print(index)
```

```python
index = string.find("Universe")
print(index)
```

NOW YOU SEE ABOVE IN THIS CODE index() function and find() function both looks very similar coz both outputs are coming same --> "7" but there is a major difference between them is in index() function when the substring not found in the output comes "substring not found" but in find() function output comes "-1"
