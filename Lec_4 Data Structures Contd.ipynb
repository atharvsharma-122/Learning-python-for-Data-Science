Join Sets

Union

```python
set1 = {1,2,3}
set2= {4,5,6}
union_set =set1.union(set2) ## set1|set2 |--> this also indicates union operation among set and set2
print(union_set)
```

Intersection

```python
set1 = {1,2,3}
set2 = {3,5,6}
intersection_set = set1.intersection(set2) ## set1& set2 &--> this indicates intersection operation among set1 and set2
print(intersection_set)
```

Difference

```python
set1 = {1,2,3}
set2 = {3,5,6}

difference_set = set1.difference(set2) ## set1 - set2 --> this indicates difference operation among set1 and set2
print(difference_set)

## it will se the elements which are present in set1 not in set2
```

Symmetric Difference

```python
set1 = {1,2,3}
set2 = {3,5,6}

symmetric_difference_set = set1.symmetric_difference(set2) ## set1 ^ set2   ^ --> this indicates symmetric difference operation among set1 and set2
print(symmetric_difference_set)

## In this common value gets dissolve and pending values display over the screen!
```

Task : Merge the below two tuples and then remove the duplicates values and return the final output as a tuple.

tup1 = (1,2,3)

tup2 = (3,4,5)

```python
tup1 = (1,2,3)
tup2 = (3,4,5)

merge_tuple = tup1 + tup2
print(merge_tuple)
```

```python
tup1 = (1,2,3)
tup2 = (3,4,5)

remove_duplicate = set(merge_tuple)
print(remove_duplicate)
```

```python
print(tuple(remove_duplicate))
```

Now your approach is also correct but we can do it more better

```python
## function definition
def merge_and_duplicates(tup1, tup2):
  return tuple(set(tup1 + tup2))

## driver code
tup1 = (1, 2, 3)
tup2 = (3, 4, 5)

## function calling
print(merge_and_duplicates(tup1, tup2))

```
