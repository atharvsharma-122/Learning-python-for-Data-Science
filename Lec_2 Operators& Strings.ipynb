**Comparison Operators**

```python
x = 5
y = 10

##Perform comparison operator
print(x == y) # Equal to
print(x != y) # Not equal to
print(x > y) # Greater than
print(x < y) # Less than
print(x >= y) # Greater than or equal to
print(x <= y) # Less than or equal to
```

**Logical Operators**

```python
x = 5
y = 10
z = 15

##Perform logical operator:\
print(x < y and y < z) # And operator
print(x < y or x > z) # Or operator
print(not(x < y)) # Not operator
```

**Bitwise Operators**

```python
x = 10
y = 6

## Perform Bitwise Operations
print(x & y) # Bitwise AND operation
print(x | y) # Bitwise OR  operation
print(x ^ y) # Bitwise XOR  operation
print(~x) # Bitwise NOT operation
print(x << 1) # Bitwise Left Shift operation
print(x >> 1) # Bitwise Right Shift operation
```

**Assignment Operators**

```python
x = 20
y = 3

## Perform assignment operators

x += y # x = x + y
print(x)

x -= y # x = x - y
print(x)

x *= y # x = x * y
print(x)

x /= y # x = x / y
print(x)

x %= y # x = x % y
print(x)

x **= y # x = x ** y
print(x)

x //= y # x = x // y
print(x)
```

**Identity Operators**

```python
x = 10
y = 10

##Perform Identity Operators
print(x is y)
print(x is not y)
```

```python
print(id(x))
print(id(y))
```
